# GCP ADK

## Hello World

```python
from google.adk.agents.llm_agent import Agent

root_agent = Agent(
    model="gemini-2.5-flash",
    name="hello_world_agent",
    description="A minimal ADK agent that answers general user queries.",
    instruction=(
        "You are a helpful, concise assistant. Answer user questions using available knowledge. "
        "If you are unsure, respond with 'I don't know' and, if possible, suggest how to find the answer."
    )
)
```

```python
from google.adk.runners import Runner, InMemoryRunner

runner = InMemoryRunner(
    agent=root_agent,
    app_name='adk_hello_world_agent'
)

while True:
    user_input = input("You: ")
    if user_input.lower() in ("exit", "quit"):
        break

    response = runner.run(
        user_id="Hello",
        session_id="123",
        new_message=user_input
    )
    print("Agent:", response.output_text)
```
