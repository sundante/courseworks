## Goal:
- Introduction to TavillySearch and basic Langchain

```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage
```

```python
load_dotenv()
```

## Very Basic

```python
llm = ChatOpenAI(model="gpt-4o")
```

```python
msg = HumanMessage(content="what is the capital of India?")
message = [msg]
```

```python
result = llm.invoke(message)
```

```python
print(result.content)
```

## Very Basic + Tavily Search

```python
from langchain_tavily import TavilySearch
```

```python
tavilysearch = TavilySearch(max_results=3)
search_result = tavilysearch.invoke("what is Langgraph?")
```

```python
search_result
```

```python
prompt = f"Summarize the following search result: {search_result}"
```

```python
result = llm.invoke(prompt)
```

```python
print(result.content)
```
