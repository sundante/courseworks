````python
"""
AMAZON CODING INTERVIEW - LAST MINUTE CHEAT SHEET
=================================================

TOP PATTERNS AMAZON LOVES (Based on real interviews!)
Practice these templates - you'll recognize variations in the interview!
"""

from collections import deque, defaultdict, Counter
import heapq

# ============================================================================
# PATTERN 1: TWO POINTERS (Very Common at Amazon!)
# ============================================================================

def two_sum_sorted(arr, target):
    """
    Two Sum in SORTED array
    Amazon LOVES this pattern!
    """
    left, right = 0, len(arr) - 1
    
    while left < right:
        current_sum = arr[left] + arr[right]
        if current_sum == target:
            return [left, right]
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    
    return [-1, -1]


def container_with_most_water(height):
    """
    LeetCode 11 - Asked at Amazon!
    Max area between two lines
    """
    left, right = 0, len(height) - 1
    max_area = 0
    
    while left < right:
        width = right - left
        max_area = max(max_area, min(height[left], height[right]) * width)
        
        # Move pointer with smaller height
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    
    return max_area


# ============================================================================
# PATTERN 2: SLIDING WINDOW (Amazon's Favorite!)
# ============================================================================

def longest_substring_no_repeat(s):
    """
    LeetCode 3 - VERY common at Amazon!
    Longest substring without repeating characters
    """
    char_index = {}
    max_length = 0
    left = 0
    
    for right in range(len(s)):
        if s[right] in char_index and char_index[s[right]] >= left:
            left = char_index[s[right]] + 1
        
        char_index[s[right]] = right
        max_length = max(max_length, right - left + 1)
    
    return max_length


def max_sum_subarray_size_k(arr, k):
    """
    Fixed-size sliding window
    Max sum of k consecutive elements
    """
    window_sum = sum(arr[:k])
    max_sum = window_sum
    
    for i in range(k, len(arr)):
        window_sum = window_sum - arr[i - k] + arr[i]
        max_sum = max(max_sum, window_sum)
    
    return max_sum


# ============================================================================
# PATTERN 3: ARRAYS & STRINGS (Bread and Butter!)
# ============================================================================

def rotate_array(arr, k):
    """
    Rotate array by k positions
    Amazon asks this in variations!
    """
    n = len(arr)
    k = k % n
    
    # Reverse entire array
    arr.reverse()
    # Reverse first k
    arr[:k] = reversed(arr[:k])
    # Reverse remaining
    arr[k:] = reversed(arr[k:])
    
    return arr


def merge_intervals(intervals):
    """
    LeetCode 56 - Asked FREQUENTLY at Amazon!
    """
    if not intervals:
        return []
    
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]
    
    for current in intervals[1:]:
        last = merged[-1]
        
        if current[0] <= last[1]:  # Overlap
            merged[-1] = [last[0], max(last[1], current[1])]
        else:
            merged.append(current)
    
    return merged


def product_except_self(nums):
    """
    LeetCode 238 - Amazon favorite!
    Product of array except self (no division!)
    """
    n = len(nums)
    result = [1] * n
    
    # Left products
    left = 1
    for i in range(n):
        result[i] = left
        left *= nums[i]
    
    # Right products
    right = 1
    for i in range(n - 1, -1, -1):
        result[i] *= right
        right *= nums[i]
    
    return result


# ============================================================================
# PATTERN 4: HASH MAPS (Use This A LOT!)
# ============================================================================

def two_sum(nums, target):
    """
    LeetCode 1 - MUST KNOW!
    """
    seen = {}
    
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    
    return [-1, -1]


def group_anagrams(strs):
    """
    LeetCode 49 - Amazon loves this!
    """
    anagrams = defaultdict(list)
    
    for word in strs:
        key = ''.join(sorted(word))
        anagrams[key].append(word)
    
    return list(anagrams.values())


# ============================================================================
# PATTERN 5: STACKS & QUEUES
# ============================================================================

def valid_parentheses(s):
    """
    LeetCode 20 - Classic Amazon question!
    """
    stack = []
    mapping = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in mapping:
            if not stack or stack[-1] != mapping[char]:
                return False
            stack.pop()
        else:
            stack.append(char)
    
    return len(stack) == 0


def daily_temperatures(temps):
    """
    LeetCode 739 - Monotonic stack pattern
    Days until warmer temperature
    """
    result = [0] * len(temps)
    stack = []  # Store indices
    
    for i, temp in enumerate(temps):
        while stack and temps[stack[-1]] < temp:
            prev_idx = stack.pop()
            result[prev_idx] = i - prev_idx
        stack.append(i)
    
    return result


# ============================================================================
# PATTERN 6: BINARY SEARCH (Quick and Powerful!)
# ============================================================================

def binary_search(arr, target):
    """
    Classic binary search - memorize this!
    """
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1


def search_rotated_array(nums, target):
    """
    LeetCode 33 - Amazon LOVES this!
    Search in rotated sorted array
    """
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        
        if nums[mid] == target:
            return mid
        
        # Left half sorted?
        if nums[left] <= nums[mid]:
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        # Right half sorted
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1
    
    return -1


# ============================================================================
# PATTERN 7: TREES (Very Common!)
# ============================================================================

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def max_depth(root):
    """Quick tree depth - memorize this!"""
    if not root:
        return 0
    return 1 + max(max_depth(root.left), max_depth(root.right))


def level_order_traversal(root):
    """
    LeetCode 102 - BFS on tree
    Amazon asks tree problems frequently!
    """
    if not root:
        return []
    
    result = []
    queue = deque([root])
    
    while queue:
        level = []
        for _ in range(len(queue)):
            node = queue.popleft()
            level.append(node.val)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        result.append(level)
    
    return result


def is_valid_bst(root):
    """
    LeetCode 98 - CRITICAL for Amazon!
    """
    def validate(node, min_val, max_val):
        if not node:
            return True
        
        if node.val <= min_val or node.val >= max_val:
            return False
        
        return (validate(node.left, min_val, node.val) and
                validate(node.right, node.val, max_val))
    
    return validate(root, float('-inf'), float('inf'))


def lowest_common_ancestor(root, p, q):
    """
    LeetCode 236 - Asked at Amazon!
    """
    if not root or root == p or root == q:
        return root
    
    left = lowest_common_ancestor(root.left, p, q)
    right = lowest_common_ancestor(root.right, p, q)
    
    if left and right:
        return root
    
    return left if left else right


# ============================================================================
# PATTERN 8: GRAPHS & BFS/DFS
# ============================================================================

def number_of_islands(grid):
    """
    LeetCode 200 - VERY POPULAR at Amazon!
    Count islands in 2D grid
    """
    if not grid:
        return 0
    
    rows, cols = len(grid), len(grid[0])
    visited = set()
    count = 0
    
    def dfs(r, c):
        if (r < 0 or r >= rows or c < 0 or c >= cols or
            (r, c) in visited or grid[r][c] == '0'):
            return
        
        visited.add((r, c))
        dfs(r + 1, c)
        dfs(r - 1, c)
        dfs(r, c + 1)
        dfs(r, c - 1)
    
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == '1' and (r, c) not in visited:
                dfs(r, c)
                count += 1
    
    return count


def shortest_path_bfs(graph, start, end):
    """
    BFS for shortest path - Amazon loves this!
    """
    if start == end:
        return 0
    
    visited = {start}
    queue = deque([(start, 0)])
    
    while queue:
        node, dist = queue.popleft()
        
        for neighbor in graph[node]:
            if neighbor == end:
                return dist + 1
            
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, dist + 1))
    
    return -1


# ============================================================================
# PATTERN 9: HEAPS/PRIORITY QUEUE (Amazon Warehouse Problems!)
# ============================================================================

def kth_largest(nums, k):
    """
    LeetCode 215 - Amazon asks this!
    Kth largest element
    """
    return heapq.nlargest(k, nums)[-1]


def top_k_frequent(nums, k):
    """
    LeetCode 347 - Top K frequent elements
    """
    count = Counter(nums)
    return [num for num, _ in count.most_common(k)]


def k_closest_points(points, k):
    """
    LeetCode 973 - Amazon delivery optimization!
    K closest points to origin
    """
    heap = []
    
    for x, y in points:
        dist = x*x + y*y
        heapq.heappush(heap, (dist, [x, y]))
    
    return [heapq.heappop(heap)[1] for _ in range(k)]


# ============================================================================
# PATTERN 10: DYNAMIC PROGRAMMING (If Time Allows)
# ============================================================================

def climbing_stairs(n):
    """
    LeetCode 70 - Easy DP starter
    """
    if n <= 2:
        return n
    
    prev, curr = 1, 2
    
    for i in range(3, n + 1):
        prev, curr = curr, prev + curr
    
    return curr


def coin_change(coins, amount):
    """
    LeetCode 322 - Classic DP
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    
    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], dp[i - coin] + 1)
    
    return dp[amount] if dp[amount] != float('inf') else -1


# ============================================================================
# QUICK REFERENCE CHEAT SHEET
# ============================================================================

def cheat_sheet():
    return """
    
    â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    ğŸš¨ AMAZON INTERVIEW QUICK REFERENCE
    â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    
    TOP 10 MUST-KNOW PROBLEMS:
    --------------------------
    1. Two Sum (LeetCode 1) â­â­â­â­â­
    2. Longest Substring Without Repeating (LeetCode 3) â­â­â­â­â­
    3. Valid Parentheses (LeetCode 20) â­â­â­â­â­
    4. Merge Intervals (LeetCode 56) â­â­â­â­â­
    5. Number of Islands (LeetCode 200) â­â­â­â­â­
    6. Validate BST (LeetCode 98) â­â­â­â­
    7. LCA Binary Tree (LeetCode 236) â­â­â­â­
    8. Product Except Self (LeetCode 238) â­â­â­â­
    9. K Closest Points (LeetCode 973) â­â­â­â­
    10. Search Rotated Array (LeetCode 33) â­â­â­
    
    
    PATTERN RECOGNITION:
    -------------------
    "Two elements sum to..."        â†’ Two Pointers / HashMap
    "Longest/shortest substring"    â†’ Sliding Window
    "K largest/smallest"            â†’ Heap
    "Maximum/minimum subarray"      â†’ Kadane's / Sliding Window
    "Valid parentheses/brackets"    â†’ Stack
    "All paths/combinations"        â†’ DFS/Backtracking
    "Shortest path"                 â†’ BFS
    "Islands/connected components"  â†’ DFS/BFS on grid
    "Interval problems"             â†’ Sort + Merge
    "Tree level/depth"              â†’ BFS (queue)
    
    
    TIME COMPLEXITY CHEAT:
    ----------------------
    O(1)      â†’ HashMap lookup, array access
    O(log n)  â†’ Binary search, heap operations
    O(n)      â†’ Single pass, sliding window
    O(n log n)â†’ Sorting, heap with n operations
    O(nÂ²)     â†’ Nested loops (try to avoid!)
    
    
    QUICK TEMPLATES:
    ----------------
    
    # Two Pointers
    left, right = 0, len(arr) - 1
    while left < right:
        # compare arr[left] and arr[right]
        # move pointers
    
    # Sliding Window
    left = 0
    for right in range(len(arr)):
        # add arr[right] to window
        while window_invalid:
            # remove arr[left] from window
            left += 1
    
    # BFS Template
    queue = deque([start])
    visited = {start}
    while queue:
        node = queue.popleft()
        for neighbor in get_neighbors(node):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    
    # DFS Template
    def dfs(node):
        if base_case:
            return
        # process node
        dfs(next_node)
    
    # Binary Search
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    
    INTERVIEW TIPS:
    ---------------
    âœ“ LISTEN carefully - understand problem first!
    âœ“ ASK questions - clarify inputs, constraints
    âœ“ DRAW examples - visualize the problem
    âœ“ START with brute force - then optimize
    âœ“ THINK OUT LOUD - explain your reasoning
    âœ“ TEST with examples - walk through your code
    âœ“ EDGE CASES - empty, single element, duplicates
    
    
    COMMON GOTCHAS:
    ---------------
    âœ— Off-by-one errors in loops
    âœ— Forgetting to handle empty input
    âœ— Integer overflow (use Python, you're safe!)
    âœ— Not considering negative numbers
    âœ— Modifying list while iterating
    âœ— Forgetting visited set in graphs
    
    
    QUICK WINS:
    -----------
    â€¢ Use defaultdict(list) for adjacency lists
    â€¢ Use Counter for frequency counts
    â€¢ Use set() for O(1) lookups
    â€¢ collections.deque for queue operations
    â€¢ heapq for priority queue
    â€¢ sorted() returns new list, .sort() in-place
    
    
    CONFIDENCE BOOSTERS:
    --------------------
    ğŸŒŸ You know more than you think!
    ğŸŒŸ It's okay to start with brute force
    ğŸŒŸ Communication > Perfect code
    ğŸŒŸ Draw pictures, talk through examples
    ğŸŒŸ Take a breath, you've got this!
    
    
    â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    """


# ============================================================================
# LAST MINUTE TIPS
# ============================================================================

if __name__ == "__main__":
    print("""
    â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
    â•‘                                                               â•‘
    â•‘         ğŸš€ AMAZON INTERVIEW - YOU'VE GOT THIS! ğŸš€            â•‘
    â•‘                                                               â•‘
    â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    
    BREATHE! You're prepared! ğŸ’ª
    
    Before you start:
    -----------------
    1. Test your audio/video
    2. Have paper and pen ready
    3. Close other apps
    4. Have water nearby
    5. Take 3 deep breaths
    
    During the interview:
    ---------------------
    â€¢ Start by repeating the problem in your own words
    â€¢ Ask about edge cases (empty input? negatives? duplicates?)
    â€¢ Draw examples before coding
    â€¢ Think out loud - they want to see your process
    â€¢ Start with brute force, then optimize
    â€¢ Test your code with examples
    
    Remember:
    ---------
    â†’ Amazon values problem-solving over perfect code
    â†’ It's okay to ask for hints
    â†’ Communication is KEY
    â†’ Explain your thought process
    â†’ They want you to succeed!
    
    You're going to do GREAT! ğŸŒŸ
    
    Good luck! Crush it! ğŸ’¯
    """)
    
    print(cheat_sheet())
````
